```python
df['names']=pd.DataFrame(['Yves', 'Sandra', 'Lilli', 'Henry'],
                         index=['d','a','b','c'])
df
```

```python
df.append({'numbers':100, 'floats':5.75, 'names':'Jil'},
          ignore_index=True)
```

```python
df=df.append(pd.DataFrame({'numbers':100, 'floats':5.75, 'names':'Jil'}, index=['y',]))
```

```python
df
```

```python
df=df.append(pd.DataFrame({'names':'Liz'}, index=['z',]),
             sort=False)
df
```

```python
df[['numbers', 'floats']].mean()
```

```python
df[['numbers', 'floats']].std()
```

### 5.1.2 DataFrame 클래스 다루기: 2단계

```python
import numpy as np

np.random.seed(100)

a=np.random.standard_normal((9,4))
a
```

```python
df=pd.DataFrame(a)
```

```python
df
```

```python
df.columns=['No1', 'No2', 'No3', 'No4']
df
```

```python
dates=pd.date_range('2019-1-1', periods=9, freq='M')
dates
```

```python
df.index=dates

df
```

```python
df.values
```

```python
np.array(df)
```
